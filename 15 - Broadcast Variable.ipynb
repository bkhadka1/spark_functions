```
transaction = [(100,'tshirt',150),
               (200,'headphone',389),
               (300,"cover",425),
               (400,"mask",789),
               (100,"jeans",890),
               (300,"perfume",987),
               (200,"cloth",234),
               (500,"bottle",678)]


tarnsactiondf = spark.createDataFrame(data=transaction,schema = ['store_id','item','amount'])
display(tarnsactiondf)
```

```
store = [(100,"goa"),
         (200,"mumbai"),
         (300,"chennai"),
         (400,"paris"),
         (500,"raipur")]

store_df =  spark.createDataFrame(data = store , schema = ["store_id","location"])
display(store_df)
```

```
from pyspark.sql.functions import broadcast

joindf = tarnsactiondf.join(broadcast(store_df),tarnsactiondf['store_id']==store_df['store_id'])

joindf.show()
```

```
joindf.explain(True)
```
